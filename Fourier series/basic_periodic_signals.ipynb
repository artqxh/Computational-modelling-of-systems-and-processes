```python
# Expand a square wave signal to a Fourier series

import numpy as np
import matplotlib.pyplot as plt

n = 25
t = np.linspace(-np.pi, np.pi, 1000);
F = 0
T = np.pi

for k in range(n):
    k += 1
    F = F + (np.sin(((2*k-1)*np.pi*t)/2*T))/(2*k-1)
    
F = F * 4/(np.pi)

plt.plot(t, F)
plt.title('square wave')
plt.show()
```

```python
# Find a square wave signal with the relative error less than 5%

import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

n = 1000
t = np.linspace(-np.pi, np.pi, 1000);
F = 0
T = np.pi

# Original wave
original_wave = signal.square(np.pi**2/2*t)

for k in range(n):
    k += 1
    F = F + 4/(np.pi)*(np.sin(((2*k-1)*np.pi*t)/2*T))/(2*k-1)
    if np.allclose(F, original_wave, rtol=5e-02) > 0:
        print('For n = {} the relative error is less than 5%'.format(k))
        break
    

plt.title('original square wave')
plt.plot(t, original_wave)
plt.show()

plt.plot(t, F)
plt.title('acceptable square wave')
plt.show()
```

```python
# Expand the triangular wave signal to a Fourier series

import numpy as np
import matplotlib.pyplot as plt

n = 25
t = np.linspace(-np.pi, np.pi, 1000);
F = 0
T = np.pi

for k in range(n):
    k += 1
    F = F + (np.cos(((2*k-1)*np.pi*t)/2*T))/(2*k-1)**2
    
F = T/2 - (4*T)/(np.pi**2)*F

plt.plot(t, F)
plt.title('triangular wave')
plt.show()
```

```python
# Expand the sawtooth wave signal to a Fourier series

import numpy as np
import matplotlib.pyplot as plt

n = 25
t = np.linspace(-np.pi, np.pi, 1000);
F = 0
T = np.pi

for k in range(n):
    k += 1
    F = F + ((2*T*(-1)**(k+1))*np.sin((k*np.pi*t)/2*T))/(k*np.pi)
    
             
plt.plot(t, F)
plt.title('sawtooth wave')
plt.show()
```
